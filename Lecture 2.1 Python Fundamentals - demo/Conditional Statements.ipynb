   - 4 types: if, if-else, if-elif-else, nested if-else

```python
number = 6
if number > 5:
    print('talentlabs')
```

```python
number = 4
if number > 5:
    print('talentlabs')
else:
    print('wrong entry')
```

```python
password = input("Enter password: ") 
if password == "/talentlabs@10": 
    print("Correct password") 
else: 
    print("Incorrect Password")
```

```python
age = 95
```

```python
if age<=20:
    print("Below 20")
elif age<=40:
    print("21-40")
elif age<=60:
    print("41-60")
else:
    print("Above 60")
```

```python
def user_check(choice): 
    if choice == 1: 
        print("Admin") 
    elif choice == 2: 
        print("Editor") 
    else: 
        print("Wrong entry")
```

```python
user_check(2)
```

```python
num = float(input("Enter a number: "))
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("Negative number")
```

#### Example 1: Write a python code to check whether an user is eligible to vote or not 
##### (Note: People with age 18+ are allowed to vote and others are not)

```python
age = int(input("Enter your age: "))

if age>=18:
    print("You are eligible to vote")
else:
    print("OOPS!! You are not allowed to vote")
```

#### Example 2: Odd or Even

```python
x = int(input("Enter the number: "))

if x%2==0:
    print(x, "is an even number")
else:
    print(x, "is an odd number")
```
